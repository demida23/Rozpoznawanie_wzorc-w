```python
import pandas as pd
import os

# Ścieżki do folderów
input_folder = './dataset_pv'  # Folder z plikami wejściowymi
output_folder = './dataset_pv_cleaned'  # Folder na poprawione pliki

# Tworzenie folderu wyjściowego, jeśli nie istnieje
os.makedirs(output_folder, exist_ok=True)

# Funkcja do czyszczenia danych
def clean_and_format_data(input_path, output_path):
    try:
        # Wczytanie danych
        data = pd.read_csv(input_path)

        # Usunięcie pustych kolumn
        data = data.dropna(axis=1, how='all')

        # Usunięcie wierszy z brakującymi wartościami
        data = data.dropna()

        # Usunięcie duplikatów
        data = data.drop_duplicates()

        # Usunięcie białych znaków z nazw kolumn
        data.columns = data.columns.str.strip()

        # Zapisanie przetworzonych danych w formacie Excel
        data.to_excel(output_path, index=False)
        print(f"Plik został zapisany: {output_path}")
    except Exception as e:
        print(f"Błąd przetwarzania pliku {input_path}: {e}")

# Iteracja po wszystkich plikach w folderze wejściowym
for file_name in os.listdir(input_folder):
    if file_name.endswith('.csv'):  # Sprawdzenie, czy plik ma rozszerzenie .csv
        input_file = os.path.join(input_folder, file_name)
        output_file = os.path.join(output_folder, file_name.replace('.csv', '_cleaned.xlsx'))
        clean_and_format_data(input_file, output_file)
```
